```python
import pandas as pd
pd.set_option('display.max_columns', None)  # 设置查看列不省略
```

```python
df = pd.read_csv('物流数据.csv')
```

```python
df['货运公司名称'].value_counts()
```

```python
df['货运公司名称'].unique()
```

```python
# 标签编码（Label Encoding）
# 假设特征取值有n个不同值，即n个类别，那么将按照特征数据的大小将其编码为0-（n-1）之间的整数。
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['le_货运公司名称'] = le.fit_transform(df['货运公司名称'])
df
```

```python
# 获取编码与类别之间的映射关系
res = {}
for cl in le.classes_:
    res.update({cl:le.transform([cl])[0]})
res
```

```python
# pd.factorize
df['fa_货运公司名称'] = pd.factorize(df['货运公司名称'], sort=True)[0]
(df['le_货运公司名称'] == df['fa_货运公司名称']).value_counts()
```

```python
# 独热编码（One-Hot Encoding）
# 哑变量
# 将分类变量转换为几个二进制列的方法，其中1表示具有该属性特征，0表示不具有该属性特征。
pd.get_dummies(df, columns=['货运公司名称'])
```

```python
# 序号编码（Ordinal Encoding）
# 对于ordinal 变量，将1到N按顺序赋值给这个N类的定序变量。类似于标签编码，但是是按照提前指定好的编号顺序对类别进行编码。
map_dict = {
    '中通速递': 0,
    '佳吉快运有限公司': 1,
    '品骏': 2,
    '圆通速递': 3,
    '天天速递': 4,
    '德邦物流': 5,
    '申通快递有限公司': 6,
    '百世汇通': 7,
    '盛辉物流': 8,
    '韵达快递': 9,
    '顺丰快递': 10
}
# df['oe_货运公司名称'] = df['货运公司名称'].replace(map_dict)  # replace，需要完全匹配，支持正则，支持根据字典替换多个值，没匹配的不会处理
df['oe_货运公司名称'] = df['货运公司名称'].map(map_dict)  # map，需要完全匹配，不能使用正则，没匹配的会被替换成NaN
(df['le_货运公司名称'] == df['oe_货运公司名称']).value_counts()  # all True
df
```

```python
# 二进制编码（Binary Encoding）
# 二进制编码将类别转换为二进制数字。每个二进制数字创建一个特征列。如果存在n个唯一类别，则二进制编码将产生log(base 2)ⁿ 个新的特征列。
# 与“独热编码”相比，这将需要较少的特征列（对于100个类别，“独热编码”需要(100-1)个特征列，而“二进制”编码仅需要七个特征，2^7=128>100）。
```

```python
import category_encoders as ce  # http://contrib.scikit-learn.org/category_encoders/index.html
encoder = ce.BinaryEncoder(cols=['货运公司名称'])
dfbin = encoder.fit_transform(df['货运公司名称'])
df = pd.concat([df, dfbin], axis=1)
df
```
